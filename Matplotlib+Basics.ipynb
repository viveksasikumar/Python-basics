```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.linspace(0,100,101)
y = x**2
```

```python
plt.plot(x,y,'b')
plt.title('Square Vs Number')
plt.xlabel('Number')
plt.ylabel('Square')
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')

plt.show()
```

```python
fig = plt.figure() #object oriented method for crating figure

axes = fig.add_axes([0.1,0.1,0.8,0.8])

axes.plot(x,y)
axes.set_xlabel('Number')
axes.set_ylabel('Square')
axes.set_title('Square Vs Number')
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.9,0.9]) #percentage from axes the size of plot
axes2 = fig.add_axes([0.2,0.6,0.4,0.3])
axes1.plot(x,y,'r',linewidth=5 ,linestyle='--',label='ABC')
axes1.set_title('Large Plot')
axes1.set_xlabel('ABC')
axes2.plot(y,x,'b',linewidth=2, linestyle='-',marker=' ',label = 'DEF')
axes2.set_title('Small Plot')
axes2.set_xlabel('DEF')
fig.legend(loc=(0.8,0.2))
```

```python
figs, axes = plt.subplots(nrows=3,ncols=3)
axes[0][1].plot(x,y)
axes[2][2].plot(y,x,'r')
```

```python
figs, axes = plt.subplots(nrows=3,ncols=3)
plt.tight_layout() #fixes the subplots spacings
```

```python
figs, axes = plt.subplots(nrows=1,ncols=2)

axes[0].plot(x,y,'r')
axes[1].plot(y,x,'b')

axes[0].annotate(r'$50*50=2500$',
             xy=(50,2500), xycoords='data',
             xytext=(-20, +30), textcoords='offset points', fontsize=10,
             arrowprops=dict(arrowstyle="->", connectionstyle="arc3,rad=.1"))

axes[0].scatter([50,],[2500,], 20, color ='grey')

axes[1].annotate(r'$6400=80*80$',
             xy=(6400,80), xycoords='data',
             xytext=(-20, +30), textcoords='offset points', fontsize=10,
             arrowprops=dict(arrowstyle="->", connectionstyle="arc3,rad=.2"))

axes[1].scatter([6400,],[80,], 20, color ='orange')
```

```python
X = np.linspace(-np.pi, np.pi, 256, endpoint=True)
C,S = np.cos(X), np.sin(X)

plt.plot(X,C)
plt.plot(X,S)

#set spines
ax = plt.gca()
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.spines['bottom'].set_position(('data',0))
ax.yaxis.set_ticks_position('left')
ax.spines['left'].set_position(('data',0))

plt.show()
```

```python
plt.plot(X,C)
plt.plot(X,S)

#annotate
t = 2*np.pi/3
plt.plot([t,t],[0,np.cos(t)], color ='blue', linewidth=1.5, linestyle="--")
plt.scatter([t,],[np.cos(t),], 50, color ='blue')

plt.annotate(r'$\sin(\frac{2\pi}{3})=\frac{\sqrt{3}}{2}$',
             xy=(t, np.sin(t)), xycoords='data',
             xytext=(+10, +30), textcoords='offset points', fontsize=16,
             arrowprops=dict(arrowstyle="->", connectionstyle="arc3,rad=.2"))

plt.plot([t,t],[0,np.sin(t)], color ='red', linewidth=1.5, linestyle="--")
plt.scatter([t,],[np.sin(t),], 50, color ='red')

plt.annotate(r'$\cos(\frac{2\pi}{3})=-\frac{1}{2}$',
             xy=(t, np.cos(t)), xycoords='data',
             xytext=(-90, -50), textcoords='offset points', fontsize=16,
             arrowprops=dict(arrowstyle="->", connectionstyle="arc3,rad=.2"))

ax = plt.gca()
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.spines['bottom'].set_position(('data',0))
ax.yaxis.set_ticks_position('left')
ax.spines['left'].set_position(('data',0))

plt.savefig('image.jpeg',dpi=300)
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, polar=True)
r = np.arange(0,1,0.001)
theta = 2*2*np.pi*r
line, = ax.plot(theta, r, color='#ee8d18', lw=3)

ind = 800
thisr, thistheta = r[ind], theta[ind]
ax.plot([thistheta], [thisr], 'o')
ax.annotate('a polar annotation',
            xy=(thistheta, thisr),  # theta, radius
            xytext=(0.05, 0.05),    # fraction, fraction
            textcoords='figure fraction',
            arrowprops=dict(facecolor='black', shrink=0.05),
            horizontalalignment='left',
            verticalalignment='bottom',
            )
plt.show()
```
